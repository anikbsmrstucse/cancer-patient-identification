## Import Liaberies

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
```

## load Dataset

```python
cancer_data = pd.read_csv('cancer-patient-data-sets.csv')
```

```python
cancer_data.head(10)
```

```python
print(cancer_data.columns)
```

## Remove Index and Paitent Id

```python
data = cancer_data.drop(columns=['index', 'Patient Id'])
```

```python
data.head(10)
```

## Find Null value

```python
null_data = data.isnull().sum()
print("Missing values:\n", null_data)
```

## Target Columns Binarey Encoding

```python
level_mapping = {
    'Low': 0,
    'Medium': 1,
    'High': 1
}
```

```python
data['Level'] = data['Level'].replace(level_mapping)
```

```python
data.head(10)
```

## Summarize the data set

```python
#dimensions(no. of rows & columns)
print(data.shape)

#list of columns/features
print(data.columns)

#statistical summary
data.describe()
```

## Specify the target variable and show its distribution

```python
#target variable
target = data['Level']

#distribution of class labels or categories
print(pd.value_counts(target))

#alternative of finding class distribution
#print(iris_data.groupby('species').size())
```

## Spilt dataset into training and test data

```python
seed = 7
train_data, test_data = train_test_split(data,test_size=0.3,random_state=seed, stratify=target)

#shape of the datasets
print('\nShape of training data :',train_data.shape)
print('\nShape of testing data : ',test_data.shape)

#class distribution of the training data
print(pd.value_counts(train_data['Level']))
#class distribution of the test data
print(pd.value_counts(test_data['Level']))
```

## Seperate the independent and target variable

```python
#Seperate the independent and target variable from the training data
train_x = train_data.drop(columns=['Level'],axis=1)
train_y = train_data['Level']

#Seperate the independent and target variables from test data
test_x = test_data.drop(columns=['Level'],axis=1)
test_y = test_data['Level']
```

## Show training data

```python
train_x.head()
```

## Show training labels

```python
train_y.head()
```

```python
from sklearn import preprocessing
normalizer = preprocessing.Normalizer()
train_x = normalizer.fit_transform(train_x)
test_x = normalizer.transform(test_x)
```

```python
train_x = preprocessing.scale(train_x)
test_x = preprocessing.scale(test_x)
```

## Save Result

```python
result = []
```

## Model build with logistic regressions

```python
#import logistics Regression
from sklearn.linear_model import LogisticRegression

#create classifier object/model
model1 = LogisticRegression()

#train the model with fit function
model1.fit(train_x,train_y)
```

## Predictions with Logistic Regression

```python
#make predictions on training data
predictions_train = model1.predict(train_x)
train_accuracy = accuracy_score(train_y,predictions_train)
print('Training Accuracy:',train_accuracy)

#make predictions on test data
predictions_test = model1.predict(test_x)
test_accuracy = accuracy_score(test_y,predictions_test)
print('Test Accuracy:', test_accuracy)

#append results
result.append(['Logistic Regression', train_accuracy, test_accuracy])
```

## Build Model with SVM

```python
#import SVM
from sklearn.svm import SVC

#create classifier object/model
model2 = SVC()

#train the model with fit function
model2.fit(train_x,train_y)
```

## Prediction with SVM

```python
#make predictions on training data
predictions_train = model2.predict(train_x)
train_accuracy = accuracy_score(train_y,predictions_train)
print('Training Accuracy:',train_accuracy)

#make predictions on test data
predictions_test = model2.predict(test_x)
test_accuracy = accuracy_score(test_y,predictions_test)
print('Test Accuracy:', test_accuracy)

#append results
result.append(['SVM', train_accuracy, test_accuracy])
```

## Build the Model(K-Nearest Neighbours)

```python
#import logistics Regression
from sklearn.neighbors import KNeighborsClassifier

#create classifier object/model
model3 = KNeighborsClassifier(n_neighbors=3)

#train the model with fit function
model3.fit(train_x,train_y)
```

## Predictions with KNN

```python
#make predictions on training data
predictions_train = model3.predict(train_x)
train_accuracy = accuracy_score(train_y,predictions_train)
print('Training Accuracy:',train_accuracy)

#make predictions on test data
predictions_test = model3.predict(test_x)
test_accuracy = accuracy_score(test_y,predictions_test)
print('Test Accuracy:', test_accuracy)

#append results
result.append(['KNN', train_accuracy, test_accuracy])
```

## compare model performances

```python
results = pd.DataFrame (result, columns=['Model','Training Accuracy','Test Accuracy'])
results
```

```python
def convert_labels_to_binary(labels, positive_label):
    return [1 if label == positive_label else 0 for label in labels]

def calculate_metrics(predictions, ground_truth):
    tp = 0
    fp = 0
    fn = 0

    for pred, label in zip(predictions, ground_truth):
        if pred == 1 and label == 1:
            tp += 1
        elif pred == 1 and label == 0:
            fp += 1
        elif pred == 0 and label == 1:
            fn += 1

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0

    return tp, fp, precision, recall
```

```python
# Example usage:
predictions = model2.predict(test_x)
ground_truth = test_y

tp, fp, precision, recall = calculate_metrics(predictions, ground_truth)

print("True Positives (TP):", tp)
print("False Positives (FP):", fp)
print("Precision:", precision)
print("Recall:", recall)
```
